##¿Qué es iterar?

Es realizar cierta acción ***varias veces***. En programación es habitual usar estructuras que permitan iterar, ya que, a diferencia de los humanos, el computador puede ejecutar una y otra vez la misma tarea sin equivocarse.

## ¿Cómo se puede iterar en python?
Python cuenta con dos estructuras bastante útiles, que son los ciclos ***for*** y ***while***. Con estos podremos simplificar las tareas.

### Ciclo for (for loop)

Por lo general, esta estructura nos permite iterar un número fijo de veces. En python nos permite recorrer secuencias, ya sean listas, tuplas, diccionarios, conjuntos y strings.

### Ciclo while (while loop)

Esta estructura nos permite iterar siempre que se cumpla cierta condición. **¡Precaución!**, se tiene que ser muy cauteloso al crear la condición, ya que si no se declara bien la condición se puede crear una ejecución infinita.

#Ciclo For

## Usando la sentencia range()

```python
#Esto nos permitirá iterar un número determinado de veces
n = 10
for i in range(n):
  print(i)
```

```python
#También puede usarse range para iterar entre un valor de inicio y uno de final
inicio = 5
final = 10
for i in range(inicio, final):
  print(i)
```

```python
#Aunque el incremento por defecto es de uno en uno, se puede implementar un incremento personalizado
incremento =-1
for i in range(final, inicio, incremento):
  print(i)
```

## Recorriendo una lista

```python
#Se puede recorrer una lista con un for, se necesita nombrar una variable que irá tomando los valores de la lista
nombres = ['Juan', 'Ana', 'Carlos', 'Pedro', 'Laura']

for nombre in nombres:
  print(f"Hola, {nombre}!")
```

## Recorriendo un diccionario

```python
#También se puede recorrer diccionarios, pero al recorrerlos nos devolverá solo las llaves
notas = {'Matemáticas': 5.0, 'Inglés': 4.0, 'Geografía': 3.5}
for jsjs in notas:
  print(f"La calificación de {jsjs} es {notas[jsjs]}")
```

##Sentencias break y continue
La sentencia ***break*** nos permite terminar el recorrido sin llegar al final del mismo, esto se hace normalmente evaluando una condición. La sentencia ***continue*** lo que hace es saltar el valor actual de la iteración y pasar directamente al siguiente

```python
#Usando la lista de nombres declarada anteriormente
print(nombres)
```

```python
#Implementamos la sentencia break
for nombre in nombres:
  #Supongamos que al saludar a Carlos, ya no queremos saludar a más nadie
  if(nombre == 'Carlos'):
    break
  print(f"Hola, {nombre}!")
else:
  print("Terminamos de saludar")
```

```python
#Implementamos la sentencia continue
for nombre in nombres:
  #Ahora supongamos que Carlos nos cae mal y no queremos saludarlo
  if(nombre == 'Carlos'):
    continue
  print(f"Hola, {nombre}!")
else:
  print("Terminamos de saludar")

#Y así ignoramos al pobre Carlitos 🙁
```

## Usando la sentencia for - else
Esta se usa cuando queremos ejecutar algo al terminar satisfactoriamente un for

```python
#Ejemplo encontrando los números primos entre 2 y 9
for n in range(2, 10):
    for x in range(2, n):
        if n % x == 0:
            print(f"{n} es igual a {x} * {n/x}")
            break
    else:
        # el ciclo se encontró sin encontrar ningún factor
        print(f'{n} Es un número primo')
```

##Implementando un while

```python
contraseña = 'holamundo'
intento = ''
tupla = (123, 'hola')
conjunto = {1,2,3}
diccionario = {
    "Clave": 10
}

#Aquí vamos a pedir la contraseña hasta que ingresen la correcta
while not intento==contraseña:
  intento = input("Ingrese la contraseña: ")
  if(intento!=contraseña):
    print("Contraseña incorrecta, intente nuevamente")
```

##Usando break

```python
errores = 0
intento = ''
while intento!=contraseña:
  intento = input("Ingrese la contraseña: ")
  if(intento!=contraseña):
    #Cada vez que haya un error, vamos incrementando esta variable
    errores+=1
    if(errores == 3):
      print("Se han cometido 3 errores, no quedan más intentos")
      #Al llegar a 3 errores usamos el break para salir del while
      break
    else:
      print("Contraseña incorrecta, intente nuevamente")
```

## Usando la sentencia while - else

```python
intento = ''
while intento!=contraseña:
  intento = input("Ingrese la contraseña: ")
  if(intento!=contraseña):
    #Cada vez que haya un error, vamos incrementando esta variable
    errores+=1
    if(errores == 3):
      print("Se han cometido 3 errores, no quedan más intentos")
      #Al llegar a 3 errores usamos el break para salir del while
      break
    else:
      print("Contraseña incorrecta, intente nuevamente")
else:
  print("Contraseña correcta, bienvenido")
```

#Serie de fibonacci con while

```python
v1 = 0
v2 = 1
n = 0
while v2<200:
  num = v1 + v2
  v1 = v2
  v2 = num
  n+=1
  print(v2)
```

#Factorial con while

```python
n = 5
resultado = n
while n > 1:
  resultado = resultado*(n-1)
  n = n-1
print(resultado)

```
